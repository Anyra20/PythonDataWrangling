The owner of the account uses special cute doggo language to describe dog stages: doggo, pupper, puppo, and floofer.
Doggolingvo slang word 'pupper' (means puppy) is used more often by WeRateDogs in twitter posts. But these posts on average are least retweeted and 'liked'. Followers are much more eager to 'like' or retweet posts with words 'doggo' or 'puppo'.



WeRateDogs really likes golden_retriever and labrador_retriever dog breeds. He posted them more often than other breeds, and his followers retweet them most too.
